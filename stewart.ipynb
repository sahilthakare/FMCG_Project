```python
nlp_cols_forecast = ['BRAND', 'ACTIVE', 'Product', 'Product description', 'PROD', 'TYP', 'STATUS']
nlp_cols_main = ['Site', 'Invoice Number', 'Customer Code', 'Name', 'Item Code', 'Item Description']
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
import string

nltk.download('stopwords')
```

```python

nlp_cols_forecast = ['Product description', 'STATUS']
nlp_cols_main = ['Name', 'Item Description']

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    if isinstance(text, str):
        text = text.lower()
        text = text.translate(str.maketrans('', '', string.punctuation))
        text = ' '.join([word for word in text.split() if word not in stop_words])
        return text
    return ""

for col in nlp_cols_forecast:
    df_forecast[col + '_processed'] = df_forecast[col].apply(preprocess_text)

for col in nlp_cols_main:
    df_main[col + '_processed'] = df_main[col].apply(preprocess_text)

tfidf_vectorizer_forecast = TfidfVectorizer(max_features=1000)
tfidf_matrix_forecast = tfidf_vectorizer_forecast.fit_transform(df_forecast[nlp_cols_forecast[0] + '_processed'] + " " + df_forecast[nlp_cols_forecast[1] + '_processed'])

tfidf_vectorizer_main = TfidfVectorizer(max_features=1000)
tfidf_matrix_main = tfidf_vectorizer_main.fit_transform(df_main[nlp_cols_main[0] + '_processed'] + " " + df_main[nlp_cols_main[1] + '_processed'])

# Store TF-IDF vectors in new columns (as lists or arrays)
df_forecast['tfidf_embeddings'] = [vec.toarray().tolist()[0] for vec in tfidf_matrix_forecast]
df_main['tfidf_embeddings'] = [vec.toarray().tolist()[0] for vec in tfidf_matrix_main]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python

def answer_question(query, df_forecast, df_main, tfidf_vectorizer_forecast, tfidf_vectorizer_main, nlp_cols_forecast, nlp_cols_main):
    processed_query = preprocess_text(query)
    query_vector_forecast = tfidf_vectorizer_forecast.transform([processed_query])
    query_vector_main = tfidf_vectorizer_main.transform([processed_query])

    forecast_embeddings = np.array(df_forecast['tfidf_embeddings'].tolist())
    main_embeddings = np.array(df_main['tfidf_embeddings'].tolist())

    if forecast_embeddings.size > 0:
        cosine_similarities_forecast = cosine_similarity(query_vector_forecast, forecast_embeddings).flatten()
        most_similar_index_forecast = cosine_similarities_forecast.argmax()
        best_match_forecast = df_forecast.iloc[most_similar_index_forecast]
        forecast_similarity_score = cosine_similarities_forecast[most_similar_index_forecast]
    else:
        best_match_forecast = None
        forecast_similarity_score = 0

    if main_embeddings.size > 0:
        cosine_similarities_main = cosine_similarity(query_vector_main, main_embeddings).flatten()
        most_similar_index_main = cosine_similarities_main.argmax()
        best_match_main = df_main.iloc[most_similar_index_main]
        main_similarity_score = cosine_similarities_main[most_similar_index_main]
    else:
        best_match_main = None
        main_similarity_score = 0

    # Determine which dataframe has a better match or if no good match is found
    if forecast_similarity_score > main_similarity_score and forecast_similarity_score > 0.1: # Threshold can be adjusted
        return "Based on forecast data: " + str(best_match_forecast)
    elif main_similarity_score > 0.1: # Threshold can be adjusted
        return "Based on main data: " + str(best_match_main)
    else:
        return "Could not find relevant information in the dataframes."

# Example usage (this part won't be executed by the AI but shows how the function works)
# query = "What is the status of HISAP15K4AF1-AE?"
# answer = answer_question(query, df_forecast, df_main, tfidf_vectorizer_forecast, tfidf_vectorizer_main, nlp_cols_forecast, nlp_cols_main)
# print(answer)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def answer_question(query, df_forecast, df_main, tfidf_vectorizer_forecast, tfidf_vectorizer_main, nlp_cols_forecast, nlp_cols_main):
    processed_query = preprocess_text(query)
    query_vector_forecast = tfidf_vectorizer_forecast.transform([processed_query])
    query_vector_main = tfidf_vectorizer_main.transform([processed_query])

    forecast_embeddings = np.array(df_forecast['tfidf_embeddings'].tolist())
    main_embeddings = np.array(df_main['tfidf_embeddings'].tolist())

    if forecast_embeddings.size > 0:
        cosine_similarities_forecast = cosine_similarity(query_vector_forecast, forecast_embeddings).flatten()
        most_similar_index_forecast = cosine_similarities_forecast.argmax()
        best_match_forecast = df_forecast.iloc[most_similar_index_forecast]
        forecast_similarity_score = cosine_similarities_forecast[most_similar_index_forecast]
    else:
        best_match_forecast = None
        forecast_similarity_score = 0

    if main_embeddings.size > 0:
        cosine_similarities_main = cosine_similarity(query_vector_main, main_embeddings).flatten()
        most_similar_index_main = cosine_similarities_main.argmax()
        best_match_main = df_main.iloc[most_similar_index_main]
        main_similarity_score = cosine_similarities_main[most_similar_index_main]
    else:
        best_match_main = None
        main_similarity_score = 0

    # Increased the similarity threshold to 0.3
    if forecast_similarity_score > main_similarity_score and forecast_similarity_score > 0.3:
        return "Based on forecast data: " + str(best_match_forecast)
    elif main_similarity_score > 0.3:
        return "Based on main data: " + str(best_match_main)
    else:
        return "Could not find relevant information in the dataframes."
```
